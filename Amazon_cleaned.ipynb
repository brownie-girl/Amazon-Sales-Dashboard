```python
import pandas as pd
```

```python
data = pd.read_csv("Amazon_sales_dataset_cleaned.csv")
```

```python
data['OrderDate']=pd.to_datetime(data['OrderDate'])
```

```python
data.head()
```

```python
#============Revenue by Category==================
```

```python
category_revenue= data.groupby('Category').agg(
    TotalRevenue = ('TotalAmount','sum'),
    OrderRevenue = ('OrderID','count')
).sort_values('TotalRevenue',ascending = False)
print("============= Revenue by Category ======================")
print(category_revenue)
```

```python
#===================Monthly sales trend=============
```

```python
data['Month']= data['OrderDate'].dt.to_period('M')
monthly_sales = data.groupby('Month')['TotalAmount'].sum()
print("\n===================Monthly sales trend=============")
print(monthly_sales)
```

```python
# Top 10 product
```

```python
data.head()
```

```python
top_product = data.groupby('ProductName').agg(
    TotalRevenue = ('TotalAmount','sum'),
    TotalQuntity = ('Quantity','sum')
).sort_values('TotalRevenue',ascending = False).head(10)
print("=================Top 10 product============================")
print(top_product)
```

```python
data.head()
```

```python
# ============ Sales by Country ==============
```

```python
Country_Sales = data.groupby('Country').agg(
    TotalRevenue =('TotalAmount','sum'),
    OrderRevenue =('OrderID','count')
).sort_values('TotalRevenue',ascending = False)
print("==================Sales by Country ==============")
print(Country_Sales)
```

```python
#================= Order Status Breakdown ==============
```

```python
Status_breakdown = data['OrderStatus'].value_counts()
Status_persent = data['OrderStatus'].value_counts(normalize=True)*100
print("\n========== Order Status Breakdown ===========")
print(pd.DataFrame({'count':Status_breakdown,'persent':Status_persent.round(2)})) 
```

```python
#=================== Payment method prefernce ====================
```

```python
data.head()
```

```python
payment_breakdown = data.groupby('PaymentMethod').agg(
    OrderCount = ('OrderID','count'),
    TotalRevenue = ('TotalAmount','sum')
).sort_values('OrderCount',ascending = False)
print("\n=========== Payment method prefernce ========")
print(payment_breakdown)
```
